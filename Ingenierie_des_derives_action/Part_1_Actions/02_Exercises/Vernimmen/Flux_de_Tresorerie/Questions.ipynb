# Questions

<details>
<summary><b> Quels sont les 4 cycles de base de l'entreprise ? </b></summary>
    
> Il y a premièrement le cycle d'exploitation , puis le cycle d'investissement, en ensuite les deux cycles de financement : cycle des capitaux propres, cycle d'endettement.

</details>

---

<details> 
    <summary>
        <b>Pourquoi dit-on que les flux de financement sont la contrepartie des flux économiques ?</b>
    </summary>

> **L'essentiel :** Le financement n'existe que pour répondre aux besoins ou à l'utilisation des surplus générés par l'activité réelle de l'entreprise. C'est le principe d'équilibre du tableau de flux de trésorerie.

Les flux économiques (somme des flux d'exploitation et d'investissement) représentent la capacité d'une entreprise à générer de la valeur. Le financement intervient comme un "miroir" :

1. **Si le flux économique est négatif (Besoin) :** L'entreprise consomme plus de cash qu'elle n'en génère. Elle doit trouver une **contrepartie** via le financement (emprunt bancaire ou augmentation de capital) pour combler ce manque.
2. **Si le flux économique est positif (Excédent) :** L'entreprise dispose de cash inutilisé. La **contrepartie** consiste alors à utiliser ce flux pour désendetter l'entreprise (remboursement) ou rémunérer les actionnaires (dividendes).

**En résumé :** Le flux de financement est la réponse "financière" au solde des opérations "réelles". Mathématiquement, l'un compense l'autre pour aboutir à la variation de trésorerie nette.

</details>
    
---

<details> 
    <summary>
        <b>Définissez l'excédent de trésorerie d'exploitation. Pouvez-vous considérer que cet excédent peut être dépensé par l'entreprise comme elle l'entend ?</b>
    </summary>

> L'excédent de trésorerie d'exploitation (ETE) est le résultat de la différence entre les dépenses de consommation courantes et les recettes perçues. $ ETE = \text{dépenses de consommation courantes} - \text{recettes perçues} $

L'excédent ne peut pas être dépensé par l'entreprise comme elle l'entend, car ces fonds peuvent servir :
1. Pour rembourser le cycle de financement (soit des dépenses plus ou moins obligatoires)
2. Lors du prochain cycle d'exploitation pour les dépenses de consommation
3. Peuvent être utilisées dans un objectif d'investissement.
    
</details>

---

<details>
    <summary>
        <b>L'excédent de trésorerie d'exploitation traduit-il en définitive un gain ?</b>
    </summary>

L'ETE prend en compte uniquement les flux de trésorerie et non : 

- La politique d'amortissement et de provisionnement
- la méthode d'évaluation des stocks
- ....

Donc on ne peut pas affirmer qu'il traduit en définitive un gain
</details>

---

<details>
    <summary>
        <b>Pourquoi dit-on qu'en règle générale, l'excédent de trésorerie d'exploitation doit être positif ? Montrer par un exemple simple que l'excédent de trésorerie d'exploitation peut être négatif en période de forte croissance, de démarrage ou d'amplitude saisonnière.</b> 
    </summary>

>En règle générale l'excédent de trésorerie doit être positif pour maintenir l'activité de l'entreprise, si on a un ETE négatif alors cela engendre un prochain cycle d'exploitation durant lequel les dépenses de consommation courantes ne pourront pas être assumées directement par les résultats de l'activité de l'entreprise alors l'entreprise brûle du cash pour exister. Dans ce cas il faut trouver une source de financement, or un financement sera accordé dans les cas où l'analyse de l'entreprise permet à l'avenir d'obtenir un ETE positif sinon les investisseurs (prêteurs) perdront les fonds investis. Plus simplement l'entreprise cessera d'exister. Il est donc important sur le long terme d'avoir des ETE positifs.

- Période de forte croissance | démarrage | amplitude saisonnière :

Dans ces types de périodes une entreprise est succeptible d'obtenir des financements importants, ou développer des solutions plus complexes ce qui peut amener à un décalage du cycle d'exploitation car doit accorder des délais de paiements à ses nouveaux clients sans avoir encore reçu le moindre argent en retour.

Pour l'amplitude saisonnière les entreprises qui font face à ces périodes doivent accumuler des stocks en avance comme par exemple les magasins de jouets à Noël.
 
</details>

<details>
    <summary>
        <b> Lorsqu'on construit un budget de trésorerie pour évaluer un investissement, les flux de trésorerie disponible peuvent-ils être négatifs ? Si Oui, est-il plus probable qu'ils le soient au début ou à l'issue du plan d'affaires ? Pourquoi ?</b>
    </summary>

> Les flux de trésorerie disponibles sont calculés sur une période donnée par : $$ \text{Flux de Trésorerie disponible} = ETE - \text{Dépenses d'investissement} $$

Lorsque le financier d'entreprise prévoit l'évolution future de l'entreprise il établit un budget de trésorerie, si dans celui-ci les dépenses d'investissements sont supérieures aux ETE alors il y aura des flux de trésorerie négatifs.

Il est plus probable que ces flux soient négatifs en début de plan d'affaires car il peut demander plus de financements et ceux-ci ont pour but d'augmenter les ETE et faire en sorte que sur le long terme les Flux de trésorerie disponibles soient positifs. Les recettes générées par l'investissement ne sont pas instantanées, l'activité met du temps pour couvrir les dépenses initiales. À l'issue du plan d'affaires les grosses dépenses initiales sont terminées et l'ETE est normalemet à son maximum grâce aux bénéfices générés par l'investissement.

</details>

<details>
    <summary>
        <b>Parmi les différents flux suivants, quels sont ceux qui vont revenir à la aux actionnaires et aux créanciers : les recettes d'exploitations, l'excédent de trésorerie d'exploitation, le flux de trésorerie disponible ? Qui en bénéficiera en priorité, les actionnaires ou les créanciers ? Pourquoi ?</b>    
    </summary>

> recettes d'exploitations : Rentrée de fond liée à l'activité de l'entreprise sur une période d'exploitation donnée. Les recettes ne reflètent pas une donnée suffisante pour savoir si lorsqu'on a payé toutes les charges obligatoires pour le déroulement de l'activité il restera ou non des fonds qui reviendront aux actionnaires et aux créanciers.

>La trésorerie d'exploitation n'a pas pour objectif d'être reversée, elle sert pour le déroulment de l'activité.

> **Les flux de trésorerie disponibles** représentent les flux qui reviendront aux actionnaire et créanciers car dans ces flux on en extrait les dépenses pour le bon déroulement de l'entreprise. Ceux qui en bénéficieront en premier sont les créanciers car dans dans les contrats, les intérêts et remboursements sont obligatoires peut importe si le flux de trésorerie disponible est positif ou négatif, il y a un engagement à respecter. Ensuite il y aura des fonds pour les actionnaires une fois que les dépenses obligatoires sont effectuées.
    
</details>

<details>
    <summary>
        <b>Un hypermarché se caractérise par une rotation des stocks de produits alimentaires très rapides (6 jours), par le paiement comptant des clients, par un important crédit de fournisseurs (45 jours), et par des frais généraux très faibles. Le cycle d'exploitation dégage-t-il un besoin de fonds ou, au contraire, un excédent de fonds ? Et une agence de voyage ? Et un distributeur de matériaux de construction ?</b>
    </summary>

1. L'hypermarché doit avoir un excédent de fonds car les recettes clients sont comptant et l'hypermarché doit donc obtenir un excédent de fonds pour ensuite payer les fournisseurs et autres charges du magasin.
2. Pour une agence de voyage, c'est le même fonctionnement, le client paie comptant lorsqu'il veut s'offir le service d'une agence de voyage qui elle ensuite avec les fonds obtenus va payer les fournisseurs (compagnie de voyage, logement, activités ...).
3. Le distributeur de matériaux de constructions va avoir tendance à recevoir des demandes qui s'effectueront dans un certain temps, la demande ne veut pas dire vente immédiate, il y a donc un délai plus grand entre la décision du client d'acheter et le paiement qu'ils recevront plus tard via devis et factures. Soit les cycles d'exploitations sont plus longs et il faut dans un premier temps un besoin de fonds afin d'avoir la matière première qui est variée en fonction des clients.
    
</details>

<details>
    <summary>
        <b>Dans une optique de trésorerie, les frais de lancement d'un nouveau parfum doivent-ils être considérés comme des dépenses d'exploitation ou des dépenses d'investissement ?</b>
    </summary>

Les frais de lancements d'un nouveau parfum sont les recherches liées à ce parfum, le marketing pour le sortir, d'autres éléments qui seront réutilisés par la suite. Ce sont des dépenses qui ont pour objectif de faire croître sur le long terme l'ETE, soit par définition cela correspond à des flux d'investissements. 

</details>

<details>
    <summary>
        <b>Comment s'analyse la décision d'investissement dans une optique de trésorerie ?</b> 
    </summary>

La décision d'investissement dans une optique de trésorerie d'analyse par rapport aux flux de trésorerie disponibles, dans le cas où il n'y aurait pas de financements précédents alors ces flux de trésorerie disponibles correspondent aux ETE, lorsque ceux-ci projetés sur plusieurs cycles montrent qu'ils sont positifs où en croissance, alors avec ces résultats ont peut plus facilement obtenir la confiance d'investisseurs ou créanciers. 

> On analyse si les FTD(ETE si dépenses d'investissements sont égales à 0) actuels et futurs sont suffisamment solide pour absorber la dépense d'investissement tout en laissant un Flux de Trésorerie Disponible sain à terme.
    
</details>
