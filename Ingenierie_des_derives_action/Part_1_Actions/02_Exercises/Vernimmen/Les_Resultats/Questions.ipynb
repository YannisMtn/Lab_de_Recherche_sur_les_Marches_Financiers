        <b>Une entreprise lève 500M€ de capitaux propres pour réaliser un projet de recherche et développement : s'est-elle enrichie ou appauvrie ? De combien ? Même question sachant qu'elle a dépensé la moitié les deux premières années et que le projet ne donnera aucun résultat. Même question sachant que la troisième année, avec le reste de ses fonds, elle acquiert une entreprise avec une prime de 25% par rapport à sa valeur, ce qui lui permettra d'améliorer son résultat cette année-là uniquement de 75M€ grâce à des synergies. </b>
    </summary>

> L'entreprise ne s'est ni enrichie, ni appauvrie, donc de 0€, les capitaux propres sont des fonds investis qui représentent un montant que les investisseurs mettent dans le but d'aider l'entreprise mais surtout avec l'objectif de se rembourser et dégager une plus value via les dividendes.

> Dans le cas où ils ont dépensé la moitié les deux premières années et qu'il n'y a toujours aucun résultat, elle s'est appauvrie de 250M€.

> Néanmoins dans le dernier cas avec le reste des fonds soit 250M€ restants l'entreprise acquiert une entreprise 25% plus chère par rapport à sa valeur :

$$\text{Prix}= \text{Valeur} \times (1+25\%)$$
$$\text{Valeur}=\frac{\text{Prix}}{(1+25\%)}$$

Soit, la valeur de l'entreprise est : 
$$\text{Valeur}=\frac{250000000}{1,25}=200000000$$

On obtient l'appauvrissement immédiat de cet achat en faisant la différence entre le montant payé et sa valeur : 
$$250M-200M=50M$$
Comme l'entreprise améliore son résultat cette année de 75M€ grâce à des synergies, alors on détermine si elle s'enrichie ou s'appauvrie en faisant la différence entre l'enrichissement et l'appauvrissement : $75M-50M=25M$

</details>

    <summary>
        <b>Une année, une entreprise a généré un flux de trésorerie disponible négatif. A-t-elle détruit ou créé de la richesse ?</b>
    </summary>

On obtient un flux de trésorerie disponible négatif lorsque : 
$$\text{Flux de Trésorerie Disponible} = \text{recettes perçues} - \text{dépenses de consommation courantes} - \text{dépenses d'investissement} \leq 0$$ 

Cela veut dire qu'on à les dépenses d'investissements et de consommations courantes qui sont supérieures aux recettes percues.

On a soit investissement exceptionnel soit des matières premières et stock supérieurs à ceux habituels :

Ce flux de trésorerie négatif crée de la richesses dans le cas où ces flux négatifs correspondent à la création de produit final qui a une valeur supérieure aux charges utilisées lors de la production.
Cependant dans le cas où ces flux négatifs correspondent à une mauvaise production et des charges surdimensionées pour un produit final qui ne vaut pas plus que les charges utilisées alors il y a une destruction de richesse. 

Dans les deux cas il y aura une destuction de richesse et une création de richesse car pour créer de la richesse il faut en détruire donc réellement on ne peut pas directement répondre à la question si elle a créé ou détruit de la richesse.

<details>
    <summary>
        <b>Analysez les points communs et les diférences entre l'optique de trésorerie et l'optique patrimoniale, en prenant pour exemple l'achat d'un terrain et l'investissement dans la recherche.</b>
    </summary>

1. **Points Communs :**
Dans les deux cas achat de terrain ou recherche, l'impact initial est identique pour la trésorerie : il y a une sortie de cash, qu'on achète un terrain ou qu'on paie des chercheurs, l'argent quitte le compte en banque. A l'instant T le flux de trésorerie est négatif.

Finalité : Les deux opérations sont des paris sur l'avenir. On débourse aujourd'hui dans l'espoir que cela serve l'activité plus tard.


2. **Différences**

L'achat d'un terrain :
    -Optique trésorerie : L'entreprise à dépensé 100 000€
    -Optique Patrimoniale : L'entreprise ne s'est pas appauvrie. Elle a simplement transformé un actif (cash) en un autre (terrain). Sa richesse totale reste inchangée.

Le terrain étant un actif non amortissable. Sauf catastrophe, sa valeur restera inscrite au bilan pour toujours au prix d'achat.

L'investissement dans la recherche : 
    -Optique trésorerie : Sortie de cash (salaires, matériel ...) et on espère que la dépense va créer un actif immatériel (brevet, savoir faire)
    -Optique patrimoniale : La recherche est considérée comme une charge, on considère que l'entreprise s'appauvrit immédiatement car on ne sait pas si la recherche va aboutir.
   
</details>

<details>
    <summary>
        <b>Le remboursement d'un emprunt apparaît-il nécessairement au compte de résultat ? En trésorerie ?</b>
    </summary>

Non sur le compte de résultat le remboursement du capital n'apparaît pas au compte de résultat, parce que rembourser un prêt n'est pas une charge qui m'appauvrit. Quand j'ai reçu les 100 000 au début par exemple, ce n'était pas un produit, c'était une dette. quand je les rends, ce n'est pas une perte c'est juste l'annulation de cette dette, je rends ce qu'on m'a prêté.

Par contre les Intêrets eux apparaissent toujours au compte de résultat, car c'est le prix du service par la banque. Ils sont classés dans les **Charges Financières**. Contrairement au capital, les intérêts représent un véritable appauvrissement.

En ce qui concerne les flux de trésorerie peu importe que ce soit l'intérêt ou le capital, si l'argent sort de la banque il apparaît dans mes flux de trésorerie.
    
</details>

<details>
    <summary>
        <b>L'augmentation de la valeur nominale d'un bien liée à l'inflation apparaît-elle au compte de résultat ?</b>
    </summary>

Non par convention comptable, sauf cas de très forte inflation. Le compte de résultat est une liste de ce que l'entreprise a gagné (**Produit**) et consommé (**Charges**) durant l'année. 

Une augmentation de valeur due à l'inflation est ce qu'on appelle un **gain latent**. Pour la comptabilité un gain n'est un produit que s'il est réalisé. Tant que l'entreprise n'a pas vendu le bien, il n'y a aucun chiffre d'affaires généré ou de produit exceptionnel. L'entreprise possède juste quelque chose qui vaut plus cher sur le marché mais cela ne rempli pas les caisses en temps normal.

**Principe de Prudence** :
Si un bien perd de la valeur, on enregistre une Dotation (une charge) pour montrer qu'on s'est appauvrit.

Si un bien gagne de la valeur (via l'inflation par exemple), on n'enregistre rien du tout, on attend que l'argent soit matérialisé par une vente pour l'inscrire en tant que profit.

</details>

<details>
    <summary>
        <b>Pourquoi l'augmentation des stocks de matière première vient-elle en diminution des achats dans le compte de résultat ?</b>
    </summary>

Parce que l'augmentation des stocks correspond à des achats stockés, puisqu'ils n'ont pas servir à la production de l'exercice, ils doivent être neutralisés pour que le compte de résultat ne reflète que la consommation. La charge de l'année ne sera représentée que par les matières premières consommées soit : 

$$\text{Charge de l'année (matière première)} = \text{Achat matières premières} + \text{stock initial de matières premières} - \text{stock final de matière premières} $$

$$ \text{Augmentation des stocks de matières première} = \text{stock final de matières premières} - \text{stock initial de matières premières} $$

</details>
