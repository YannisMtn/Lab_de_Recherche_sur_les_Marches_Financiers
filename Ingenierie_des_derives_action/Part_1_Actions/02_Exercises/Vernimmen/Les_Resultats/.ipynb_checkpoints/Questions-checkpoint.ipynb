<details>
    <summary>
        <b>Une entreprise lève 500M€ de capitaux propres pour réaliser un projet de recherche et développement : s'est-elle enrichie ou appauvrie ? De combien ? Même question sachant qu'elle a dépensé la moitié les deux premières années et que le projet ne donnera aucun résultat. Même question sachant que la troisième année, avec le reste de ses fonds, elle acquiert une entreprise avec une prime de 25% par rapport à sa valeur, ce qui lui permettra d'améliorer son résultat cette année-là uniquement de 75M€ grâce à des synergies. </b>
    </summary>

> L'entreprise ne s'est ni enrichie, ni appauvrie, donc de 0€, les capitaux propres sont des fonds investis qui représentent un montant que les investisseurs mettent dans le but d'aider l'entreprise mais surtout avec l'objectif de se rembourser et dégager une plus value via les dividendes.

> Dans le cas où ils ont dépensé la moitié les deux premières années et qu'il n'y a toujours aucun résultat, elle s'est appauvrie de 250M€.

> Néanmoins dans le dernier cas avec le reste des fonds soit 250M€ restants l'entreprise acquiert une entreprise 25% plus chère par rapport à sa valeur :

$$\text{Prix}= \text{Valeur} \times (1+25\%)$$
$$\text{Valeur}=\frac{\text{Prix}}{(1+25\%)}$$

Soit, la valeur de l'entreprise est : 
$$\text{Valeur}=\frac{250000000}{1,25}=200000000$$

On obtient l'appauvrissement immédiat de cet achat en faisant la différence entre le montant payé et sa valeur : 
$$250M-200M=50M$$
Comme l'entreprise améliore son résultat cette année de 75M€ grâce à des synergies, alors on détermine si elle s'enrichie ou s'appauvrie en faisant la différence entre l'enrichissement et l'appauvrissement : $75M-50M=25M$

</details>

<details>
    <summary>
        <b>Concrètement, en reprenant le schéma du paragraphe 3.15, de quoi l'entreprise s'enrichit-elle au cours d'un exercice donné ? Pourquoi</b>
    </summary>

> Le Schéma du paragraphe 3.15 résume les approches de présentation du compte de résultat (par nature | par fonction).

Concrètement l'entreprise s'enrichit lorsqu'elle obtient un résultat net positif. Soit lorsque : 
$$\text{Résultat d'exploitation} - \text{Charges financières nettes des produits financiers} + \text{Résultat non récurrent} - \text{Impôt sur les sociétés} \geq 0$$
    
</details>

<details>
    <summary>
        <b>Commentez les deux phrases suivantes : "Cette année, nous allons devoir nous endetter pour combler nos pertes"; "le résultat que nous avons dégagé cette année va nous permettre de recheter notre principal concurrent".</b>
    </summary>

1. Cette phrase confond l'objectif de l'endettement, il n'augmente pas la valeur de l'entreprise et doit être remboursé, si cet endettement à pour objectif de combler des pertes et non à proposer des solutions dans le but d'augmenter les resultats net de l'entreprise l'objectif n'est pas le bon. Néanmoins dans le sens de trésorerie, les pertes peuvent demander des besoins de trésorerie que l'endettement comble.
2. Le résultat dégagé pour l'entreprise ne constitue pas le moyen de financement d'un investissement, le résultat est une notion comptable virtuelle, c'est un indicateur d'enrichissement sur l'année, mais il peut y avoir un décalage de paiement, soit avoir un résultat de 100M€ mais ne rien avoir en banque. Il prend en compte aussi des charges non décaissées pour le moment, on achète pas une entreprise avec un résultat mais avec du cash. Il faut donc regarder au niveau de la trésorerie pour regarder les moyens de financement. Cela peut être le cas cependant lorsqu'il n'y a aucun décalage et que les flux de trésorerie net du même montant donc il n'y a pas de délais que ce soit pour les fournisseurs ou les clients, ce qui est très rare.
</details>

<details>
    <summary>
        <b>Une année, une entreprise a généré un flux de trésorerie disponible négatif. A-t-elle détruit ou créé de la richesse ?</b>
    </summary>

On obtient un flux de trésorerie disponible négatif lorsque : 
$$\text{Flux de Trésorerie Disponible} = \text{recettes perçues} - \text{dépenses de consommation courantes} - \text{dépenses d'investissement} \leq 0$$ 

Cela veut dire qu'on à les dépenses d'investissements et de consommations courantes qui sont supérieures aux recettes percues.

On a soit investissement exceptionnel soit des matières premières et stock supérieurs à ceux habituels :

Ce flux de trésorerie négatif crée de la richesses dans le cas où ces flux négatifs correspondent à la création de produit final qui a une valeur supérieure aux charges utilisées lors de la production.
Cependant dans le cas où ces flux négatifs correspondent à une mauvaise production et des charges surdimensionées pour un produit final qui ne vaut pas plus que les charges utilisées alors il y a une destruction de richesse. 

Dans les deux cas il y aura une destuction de richesse et une création de richesse car pour créer de la richesse il faut en détruire donc réellement on ne peut pas directement répondre à la question si elle a créé ou détruit de la richesse.
</details>

<details>
    <summary>
        <b>Un Excédent Brut d'exploitation apparaît-il nécessairement en trésorerie lorsqu'il est constraté comptablement ?</b>
    </summary>

>Non, l'EBE prend en compte les échanges futurs, livraisons/facturations qui correspondent au cycle d'exploitation actuel, mais cela ne veut pas dire que que des flux liés à cette exploitation ont déja eu lieu, par exemple les dépenses d'exploitations au fournisseur se font peut être à la fin de l'exploitation et ne sont pas encore présents dans les flux de trésorerie. Idem pour les paiements des clients qui ne sont peut être pas encore présents dans les flux mais comptabilisés car une facturation à la date actuelle est effectuée.

En comptabilité l'EBE est calculé selon la comtabilité d'engagment, dès que la prestation est faite ou le produit livré on enregistre la vente, même si le client à 60 jours pour payer.

</details>
