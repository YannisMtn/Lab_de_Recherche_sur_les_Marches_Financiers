<details>
    <summary>
        <b>À quoi est égale une variation de "stocks"</b>
    </summary>
Les stocks sont composés de : Produits finis ou de matières premières. Une variation signifie soit qu'une matière première a été consommée car le stock baisse, donc création d'un produit fini dans le but d'être vendu, soit que cette matière première vient d'être achetée et le stock monte. 
Dans le cas de la variation de stock de produit finis, cela implique soit qu'un produit finis est ajouté au stock en attente d'être vendu, soit qu'il y en a en moins donc une vente. Cela génère dans chacun de ces cas un flux.
</details>

    </summary>
Un cumul de flux est égal à un stock.
</details>

<details>
    <summary>
        <b>La société publiant le magazine Tour du monde du chapitre 3 (exercice 3)
a-t-elle un besoin en fonds de roulement positif ou négatif ?</b>
    </summary>
Pour calculer le BFR je rappelle de quoi il est composé : 
$$\text{BFR}= \text{Stocks} + \text{Créances d'exploitation} - \text{Dettes d'exploitation}$$

Il y a une absence de créances clients (paiement comptant à l'abonnement), une absence de stock et l'existence de dettes d'exploitation importantes sous forme de produits constatés d'avance. L'entreprise encaisse le prix de 12 magazines avant même d'avoir payé les frais de fabrication des 11 suivants.

>Donc comme les seules données non nulles sont les dettes d'exploitation, alors le BFR est négatif, on parle d'excédent en fond de roulement.

</details>

    <summary>
        <b>Classez les éléments suivants en stock, flux, ou variation de flux : chiffre d’affaires, compte clients, variation du compte clients, accroissement de dividendes, frais financiers, accroissement du chiffre d’affaires, excédent de trésorerie d’exploitation.</b>
    </summary>

Le Chiffre d'affaires correspond à des flux. C'est le cumul des ventes sur une période donnée (un exercice). C'est le flux d'activité principal.

Le compte clients correspond à un stock, c'est une créance, c'est-à-dire un stock de factures non encore encaissées.

variation de compte client est un flux. C'est la différence entre le stock final et le stock initial. C'est ce flux qui explique pourquoi le cash encaissé est différent du CA.

Accroissement de dividendes est une variation de flux. C'est la différence entre le flux de dividendes de l'année N et celui de l'année N-1.

Frais financiers est un flux. C'est une charge de l'exercice, donc un flux (consommation de ressources durant la période).

Accroissement du chiffre d'affaires est une variation de flux.C'est le delta entre deux flux de CA. C'est ce qu'on appelle souvent la croissance organique

ETE = flux, c'est le flux de trésorerie réel généré par l'exploitation.

</details>

<details>
    <summary>
        <b>Les ventes de l’entreprise sont indéniablement une ressource. Pourtant, elles
n’apparaissent pas dans le bilan économique. Pourquoi ?</b>
    </summary>

Le bilan est un document qui recense les stocks de richesse (actifs et passifs) à un instant donné, tandis que les ventes sont des flux de revenus.

Les ventes n'apparaissent pas au bilan car elles ne constituent pas une valeur détenue, mais une mesure de l'activité. Seul le solde net de ces ventes (le résultat), après déduction de toutes les charges, est inscrit au bilan dans les capitaux propres. Les ventes sont déjà 'transformées' au bilan : soit en trésorerie (si le client a payé), soit en créances (s'il n'a pas encore payé).
</details>

- Immobilisations : **Marque** (C'est une immobilisation incorporelle. Elle a vocation à rester durablement dans l'entreprise pour générer du CA), **Titres de filiales**(Ce sont des immobilisations financières (Titres de Participation). L'entreprise compte les garder à long terme pour exercer un contrôle ou une influence sur une autre société.)
- BFR : **Stock de produits finis** (C'est un emploi d'exploitation. L'argent est "bloqué" dans l'attente de la vente. C'est le cœur du cycle court.), **Dette Fiscale** (C'est une ressource d'exploitation (ou hors exploitation). L'entreprise a une dette envers l'État (TVA à payer, IS). Tant qu'elle n'est pas payée, c'est de la trésorerie que l'entreprise conserve "gratuitement". Elle vient donc réduire le BFR),**Créance de TVA**(C'est un emploi hors exploitation. C'est de l'argent que l'État doit rembourser à l'entreprise. En attendant le remboursement, cet argent est "dehors", il augmente donc le besoin de financement (BFR)) 
- Capitaux propres :**Résultat mis en réserve** (C'est de l'autofinancement. Une fois le bénéfice constaté, l'Assemblée Générale décide de ne pas le distribuer),

Il vient s'ajouter à l'endettement brut pour définir la position nette de l'entreprise vis-à-vis des banques.), **emprunts obligataires** (C'est une dette financière stable (moyen/long terme). Elle est contractée auprès d'investisseurs sur les marchés pour financer l'actif économique. Elle constitue la part brute de l'endettement.),

<details>
    <summary>
        <b>Une entreprise en cessation de paiement est-elle obligatoirement
insolvable ?</b>
    </summary>
On mesure la solvabilité d'une entreprise lorsque l'on met fin à l'exploitation de l'entreprise et liquide tout ses actifs pour faire face aux engagments(passifs). Dans le cas où elle ne peut pas alors on évalue que la différence entre actif et passif donne un résultat négatif et conclu que l'entreprise est insolvable. Cependant une cessation de paiement ne qualifie pas forcément l'entreprise d'insolvable car s'il y a une cessation de paiement cela peut indiquer une illiquidité de l'entreprise, soit qu'elle ne possède pas de fonds disponibles à court terme et ne peux donc pas faire face à ses engagements.
    
</details>

<details>
    <summary>
        <b>Dans quel cas utilise-t-on le bilan dans une optique patrimoniale ? Le bilan dans une optique économique ?</b>
    </summary>

Le bilan patrimonial à pour but de répondre à la question si on vend tout aujourd’hui que resterait-il vraiment dans la poche des propriétaires, donc dans le cas d'une étude de l'entreprise par des analystes, gestionnaires de patrimoine pour évaluer le risque qu'une institution doit faire face si elle travaille en collaboration avec l'entreprise évaluée.

Le bilan économique, tient compte des règles de prudence et vient étudier la richesse de l'entreprise sur l'exercice et ne prend donc pas en compte les éléments non traités habituellement dans le cycle d'exploitation s'il n'y a pas eu de cession sur ceux-ci. On va donc comprendre les besoins réels de l'entreprise.

</details>

<details>
    <summary>
        <b>Quelle est la différence entre des engagements et des ressources ?</b>
    </summary>

Les engagements représentent ce que l'entreprise doit contractuellement à des créanciers, clients, organismes envers lesquels l'entreprise doit des fonds ou produits. Les ressources se divisent en deux catégories :  les capitaux propres (qui ne sont pas exigibles à une date fixée) et aussi les engagements. 
    
</details>

<details>
    <summary>
        <b>Citez des activités où le besoin en fonds de roulement est positif.</b>
    </summary>

Le BFR est positif dans les activités où le cycle de production est long (stocks importants) et où les clients paient à échéance (créances clients). C'est le cas de l'industrie (Automobile) ou aéronautique. Le crédit accordé par les fournisseurs ne suffit pas à compenser ces emplois, ce qui crée un besoin de financement permanent.
    
</details>

---

<details>
    <summary>
        <b>Citez les activités où le besoind en fonds de roulement est négatif.</b>
    </summary>
À l'inverse du cas précédent, cela va être lorsque les entreprises possèdent des stocks rapides ou non existants et ayant des délais de dettes d'exploitation supérieurs au délais de paiment tel que les clients paient comptant ou sans un long délai de paiement. Par exemple les épiceries alimentaires possèdent des dettes fournisseur (dettes d'exploitation) qui sont compensées par le fait que les stocks sont rapides et que les paiements clients sont comptant, ce qui donne un BFR négatif.Les activités de services à abonnement : l'encaissement a lieu avant la réalisation de la prestation (Produits constatés d'avance), créant une ressource d'exploitation structurelle.
    
Dans ces cas, le BFR devient une ressource de financement au lieu d'être un besoin.

</details>

<details>
    <summary>
        <b>Le bilan en lecture économique fait-il une différence entre les dettes bancaires à plus d’un an et celles à moins d’un an ? Pourquoi ?</b>  
    </summary>
Le bilan en lecture économique ne distingue pas les dettes bancaires selon leur échéance (plus ou moins d'un an).

La raison est que l'analyse économique cherche à mesurer l'Actif Économique et la manière dont il est financé globalement. Toutes les dettes bancaires sont regroupées dans l'Endettement Financier Net. On considère que la distinction court/long terme est une question de gestion de trésorerie, mais que sur le plan du financement de l'outil de travail, elles ont la même fonction
    
</details>
