<details>
    <summary>
        <b>Les méthodes d'évaluation des stocks exercent-elles une influence : 
            
        - sur les résultats de l'entreprise ?
        - sur la situation de trésorerie de l'entreprise ?
</b>
</summary>

La méthode d'évaluation des stocks exerce une influence sur les résultats de l'entreprise car s'ils sont sur ou sous valorisés, le résultat immédiat sera soit plus élevé soit plus bas que celui valorisé correctement ce qui influera sur le résultat suivant, si il était sur évalué à l'origine le suivant sera plus faible et inversement s'il était sous évalué.

Non, il n'y a pas d'influence sur la trésorerie de l'entreprise.
    
</details>

---

<details>
    <summary>
        <b>Quelles sont les différences entre les flux de trésorerie provenant de l'exploitation et l'ETE ?</b>
    </summary>

Les flux de trésorerie provenant de l'exploitation incluent les frais financiers, l'impôt et certaines opérations exceptionnelles. Puisqu'ils sont établis en partant du résultat net sans que ces éléments soient corrigés.
</details>

---

<details>
    <summary>
        <b>Quelles sont les charges calculées qui doivent être réintégrées dans le calcul de la capacité d'autofinancement ?</b>
    </summary>

il faut y intégrer en partant du résultat net, la dotation aux amortissements, dépréciation d'actifs immobilisés, provisions à caractère de réserve.
    
</details>

---

<details>
    <summary>
        <b>La capacité d'autofinancement est-elle une mesure d'accroissement de richesse ? de trésorerie ?</b>
    </summary>

Non car la dotation aux amortissements traduit une usure des immobilisations et donc un appauvrissement qui n'est pas pris en compte dans la capacité d'autofinancement.

Non car les clients ne paient pas comptant et car les fournisseurs ne sont pas payés comptant.

</details>

<details>
    <summary>
        <b>Pourquoi la différence entre l'EBE et l'ETE correspond-elle à l'accroissement du besoin en fond de roulement ?</b>
    </summary>

L'EBE (Excédent Brut d'Exploitation) : C'est le profit généré par l'exploitation avant de payer les impôts, les intérêts et avant de compter l'usure des machines. C'est un flux de comptabilité d'engagement : on compte une vente dès que la facture est émise, même si le client ne paiera que dans 3 mois.

L'ETE (Excédent de Trésorerie d'Exploitation) : C'est le cash réellement encaissé par l'entreprise grâce à son activité. C'est un flux de comptabilité de trésorerie.

La différence entre ce que l'entreprise a vendu (EBE) et ce qu'elle a encaissé (ETE) est "piégée" dans le cycle d'exploitation. C'est là qu'intervient le Besoin en Fonds de Roulement (BFR).

Si le BFR augmente, cela signifie que :

Les Stocks ont augmenté
Les Créances Clients ont augmenté
Les Dettes Fournisseurs ont diminué

La différence entre l'EBE et l'ETE correspond à l'accroissement du BFR car c'est le montant des profits qui n'a pas encore pu se transformer en monnaie sur le compte à cause des décalages de paiement et des stocks.
    
</details>

---

<details>
    <summary>
        <b>Quelle est la différence entre capacité d'autofinancement et flux de trésorerie provenant de l'exploitation ?</b>
    </summary>

La CAF représente l'ensemble des ressources internes générées par l'entreprise pour financer son activité. On l'appelle aussi parfois "Cash-flow brut".
Ce qu'elle mesure : La richesse créée par l'activité une fois qu'on a payé les charges décaissables (salaires, impôts, loyers, intérêts). Elle ignore les décalages de paiement. La CAF suppose que tout ce qui a été vendu a été encaissé, et que tout ce qui a été acheté a été payé. C'est un flux avant variation du Besoin en Fonds de Roulement (BFR).

Le FTE (ou Operating Cash Flow) est ce qui reste réellement dans la caisse de l'entreprise après avoir pris en compte le "poids" du cycle d'exploitation.
Ce qu'il mesure : L'argent liquide qui est effectivement entré ou sorti pendant l'année à cause de l'exploitation. Il prend en compte le fait que les clients paient tard ou que les stocks augmentent.

Le passage de la CAF au Flux de Trésorerie d'Exploitation se fait par la variation du BFR : $$\text{Flux de Trésorerie d’Exploitation= CAF − ΔBFR}$$

Si le BFR augmente (ΔBFR>0) : L'entreprise a consommé du cash pour financer ses clients ou ses stocks. Le flux de trésorerie sera inférieur à la CAF.

Si le BFR baisse (ΔBFR<0) : L'entreprise a dégagé du cash (en vendant ses stocks ou en étant payée plus vite). Le flux de trésorerie sera supérieur à la CAF.
    
</details>

<details>
    <summary>
        <b>
Vrai ou faux : 
            
a) L’EBE est un ETE potentiel ;
            
b) Le compte de résultat repose sur une approche patrimoniale ;

c) L’entreprise s’enrichit chaque année de son EBE ;

d) Le remboursement d’un emprunt apparaît au compte de résultat ;

e) Les méthodes d’évaluation des stocks exercent une influence directe sur la
trésorerie ;

f) Les hypermarchés ont généralement des BFR négatifs ;

g) L’ETE est, en général, inférieur à l’EBE ;

h) La différence entre l’EBE et l’ETE d’une période donnée correspond au

montant du BFR ;
i) Le désendettement net de l’entreprise provient de reprises de dépréciations ;

j) Toutes choses égales par ailleurs, un accroissement du BFR entraîne un
désendettement net.</b>
    </summary>

a) Vrai

b) Vrai

c) Faux, d'autres charges viennent en déduction de l'EBE avant d'atteindre le résultat net

d) Faux, ce n'est pas une charge ou un appauvrissement, cependant les intérêts y sont.

e) Faux, la valeur à laquelle on décide de valoriser un stock n'a aucune relation avec les liquidités que reçoit ou débourse l'entreprise

f) Vrai

g) Vrai

h) Faux, c'est la variation du BFR

i) Faux, pas d'impact en trésorerie, à la fiscalité près

j) Faux, au contraire c'est une augmentation de l'endettement net.

    
</details>

<details>
    <summary>
        <b>Pourquoi, dans la capacité d’autofinancement, les plus-values de cession viennent-elles en déduction ?</b>
    </summary>

Pour être extournées dans la détermination de la capacité d'autofinancement parce qu'elles sont incluses dans le résultat net. Et il est nécessaire de les extourner pour éviter de les compter deux fois car elles apparaissent déja dans le tableau de flux dans le prix de cession des actifs cédés.

    
</details>

    <summary>
        <b>Une augmentation de capital par incorporation de réserves apparaît-elle dans le tableau de flux ?</b>

Non cette opération n'a pas d'impacte en trésorerie.
    
</details>
